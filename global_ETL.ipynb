DATA CLEANING AND TRANSFORMATION

```python
### ### There is no mssing value as the number of rows are same but the order_date is supposed to bein datetime data type 
# not string object, thus we will convert the data type of Order_Date to datetime.

global_df['Order_Date'] = pd.to_datetime(global_df['Order_Date'])
```

```python
global_df.info()
```

DATA MODELLING / NORMALZATION

```python
##We will first build tables with primary keys then build tables with foreign keys.

#Customer Table

customer_df = global_df[['Customer_ID', 'Customer_Location']].copy().drop_duplicates().reset_index(drop=True)
customer_df.head(30)
```

```python
#Product Table

product_df = global_df[['Product_Category', 'Product_Sub_Category', 'Price', 'Discount']].copy().drop_duplicates().reset_index(drop=True)

product_df.head(30)
```

```python
#Order Table
Order_df = global_df[['Order_ID', 'Customer_ID', 'Product_Category', 'Order_Date', 'Quantity', 'Revenue']].copy().drop_duplicates().reset_index(drop=True)
Order_df.head(30)
```

```python
#Save the tables as csv files

customer_df.to_csv(r'dataset\cleaneddata\customer_table.csv', index=False)
Order_df.to_csv(r'dataset\cleaneddata\order_table.csv', index=False)
product_df.to_csv(r'dataset\cleaneddata\product_table.csv', index=False)
```
